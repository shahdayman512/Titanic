```python
Age_mean  = df['Age'].mean()
```

```python
df['Age'].fillna(Age_mean, inplace=True)
```

```python
df['Cabin'].fillna('unknown', inplace=True)
```

```python
df['Fare'].fillna(0, inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['Floor']=df['Cabin'].str[0]
df
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
categorical_columns
```

```python
hot_one= pd.get_dummies(df['Sex'], prefix='Sex')
df1=pd.concat([df, hot_one], axis=1)
df1=df1.drop('Sex', axis=1)
df1
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le=LabelEncoder()
df1['Embarked_lapel']=le.fit_transform(df1['Embarked'])
df1=df1.drop('Embarked', axis=1)
df1
```

```python
Age=df1['Age'].astype(int)
```

```python
 age_count =Age.value_counts()
 age_count
```

```python
xpoints = age_count.index
ypoints = age_count.values

plt.plot(xpoints, ypoints , marker = 'o' ,mfc = 'r')
plt.title("number of passengers across different ages")
plt.xlabel("Age")
plt.ylabel("Count of passengers")
plt.show()
```

```python
pclass_count =df1['Pclass'].value_counts()
pclass_count
```

```python
x = pclass_count.index
y = pclass_count.values

font1 = {'family':'serif','color':'indigo','size':13}
font2 = {'family':'serif','color':'mediumorchid','size':10}

plt.title("Compare the number of passengers for each Pclass", fontdict = font1)
plt.xlabel("Passenger Class", fontdict = font2)
plt.ylabel("Count of passengers", fontdict = font2)

plt.bar(x,y, color = 'plum', width = 0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
x=df1['Age']
y=df1['Fare']
colors = {0: 'orange', 1: 'brown'}

plt.scatter(x, y, c=df1['Survived'].map(colors))

plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(['dead', 'Survived'])


plt.show()
```

```python
 Survived_count =df['Survived'].value_counts()
 Survived_count
```

```python
y=Survived_count.values
mylabels=['Died','Survived']
colors=['lightseagreen','turquoise']
font3 = {'family':'serif','color':'teal','size':17}
plt.title("Survived vs Died", fontdict = font3)
plt.pie(y,labels=mylabels, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
survived_gender = df.groupby(['Sex', 'Survived']).size().unstack()
survived_gender
```

```python
survived_gender.plot(kind='bar', stacked=True, color=['salmon', 'skyblue'])
plt.title('Survival Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['Did not survive', 'Survived'])
plt.show()
```
